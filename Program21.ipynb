```python
#program for calculating discount


#taking the input from the user
def discount_calc(price, discount):
    discount_amt = price * discount / 100
    return price - discount_amt

price = float(input("Enter price: "))
discount = float(input("Enter discount %: "))

print("Final price:", discount_calc(price, discount))



```
