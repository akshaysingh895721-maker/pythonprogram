```python
#program to calculate electricity bill


#taking the input from the user
units = int(input("Enter units: "))

slabs = {100:5, 200:7}  # rates per unit
if units <= 100:
    bill = units * slabs[100]
elif units <= 200:
    bill = 100*slabs[100] + (units-100)*slabs[200]
else:
    bill = 100*slabs[100] + 100*slabs[200] + (units-200)*10

print("Total bill:", bill)







```
